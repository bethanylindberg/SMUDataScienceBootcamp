# PyCity Schools Analysis

```python
# Dependencies and Setup
import pandas as pd
from pandas import DataFrame

school_data_to_load = pd.read_csv('C:\\Users\\bethf\\Desktop\\Resources\\schools.csv')
student_data_to_load = pd.read_csv('C:\\Users\\bethf\\Desktop\\Resources\\students.csv')

# Read School and Student Data File and store into Pandas Data Frames
school_data = (school_data_to_load)
student_data = (student_data_to_load)

# Combine the data into a single dataset
schooldatacomplete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])

schooldatacomplete = schooldatacomplete.rename(columns={"school_name": "School Name","type":"Type"})

schooldatacomplete.head()
```

## District Summary

```python
#Calculations to find values for dataframe
totalschools = schooldatacomplete["School Name"].nunique() #Unique school name to get count of schools
totalstudents = schooldatacomplete["Student ID"].nunique() #Unique student IDs to get count of students
totalbudget = school_data["budget"].sum() #Sum of budget for each school
averagemathscore = round(schooldatacomplete["math_score"].mean(),2) #Average math score for entire district
averagereadingscore = round(schooldatacomplete["reading_score"].mean(),2) #Average reading score for entire district
#Scores greater than 70 divided by student count to get passing rate for entire district
schoolspassingmath = (schooldatacomplete.loc[schooldatacomplete["math_score"]>=70,"math_score"].count()/totalstudents)*100
schoolspassingreading = (schooldatacomplete.loc[schooldatacomplete["reading_score"]>=70,"reading_score"].count()/totalstudents)*100
schoolspassingoverall = (averagemathscore + averagereadingscore) / 2

#Construct dataframe for values
district_summary_table = DataFrame({"Total Schools":totalschools,\
                                    "Total Students":totalstudents,\
                                    "Total Budget":totalbudget,\
                                    "Average Math Score":averagemathscore,\
                                    "Average Reading Score":averagereadingscore,\
                                    "% Passing Math":schoolspassingmath,\
                                    "% Passing Reading":schoolspassingreading,\
                                    "% Overall Passing Rate":schoolspassingoverall},index=[0])

#Summary table formatting
district_summary_table["Total Students"] = district_summary_table["Total Students"].map("{:,.0f}".format)
district_summary_table["Total Budget"] = district_summary_table["Total Budget"].map("${:,.2f}".format)
district_summary_table["% Passing Math"] = district_summary_table["% Passing Math"].map("{:.2f}%".format)
district_summary_table["% Passing Reading"] = district_summary_table["% Passing Reading"].map("{:.2f}%".format)
district_summary_table["% Overall Passing Rate"] = district_summary_table["% Overall Passing Rate"].map("{:.2f}%".format)

district_summary_table
```

## School Summary

```python
#Above caluculations grouped by school
studentsperschool = schooldatacomplete.groupby(["School Name","Type"])["Student ID"].count()
schoolbudget = schooldatacomplete.groupby(["School Name","Type"])["budget"].max()
mathpassingper = round(schooldatacomplete[schooldatacomplete["math_score"]>=70].groupby("School Name")["Student ID"].count()/studentsperschool*100,2)
readingpassingper = round(schooldatacomplete[schooldatacomplete["reading_score"]>=70].groupby("School Name")["Student ID"].count()/studentsperschool*100,2)

school_summary = DataFrame({"Total Students":studentsperschool,\
                            "Total School Budget":schoolbudget,\
                            "Per Student Budget":schoolbudget/studentsperschool,\
                            "Average Math Score":round(schooldatacomplete.groupby(["School Name","Type"])["math_score"].mean(),2),\
                            "Average Reading Score":round(schooldatacomplete.groupby(["School Name","Type"])["reading_score"].mean(),2),\
                            "% Passing Math":mathpassingper,\
                            "% Passing Reading":readingpassingper,\
                            "% Overall Passing Rate":round((mathpassingper + readingpassingper)/2,2)})

#Summary table formatting
school_summary["Total School Budget"] = school_summary["Total School Budget"].map("${:,.2f}".format)

school_summary
```

## Top Performing Schools (By Passing Rate)

```python
#Above table sorted to show top performing schools
top_performing_schools = school_summary.sort_values("% Overall Passing Rate",ascending =False)

top_performing_schools
```

## Bottom Performing Schools (By Passing Rate)

```python
#Above table sorted to show bottom performing schools
bottom_performing_schools = school_summary.sort_values("% Overall Passing Rate")

bottom_performing_schools.head()
```

## Math Scores by Grade

```python
#Math scores extracted by grade
grade9math = schooldatacomplete.loc[schooldatacomplete["grade"]=="9th","grade":"math_score"]
grade10math = schooldatacomplete.loc[schooldatacomplete["grade"]=="10th","grade":"math_score"]
grade11math = schooldatacomplete.loc[schooldatacomplete["grade"]=="11th","grade":"math_score"]
grade12math = schooldatacomplete.loc[schooldatacomplete["grade"]=="12th","grade":"math_score"]

math_scores_by_grade = DataFrame({"9th":round(grade9math.groupby(["School Name"])["math_score"].mean(),2),\
                                  "10th":round(grade10math.groupby(["School Name"])["math_score"].mean(),2),\
                                  "11th":round(grade11math.groupby(["School Name"])["math_score"].mean(),2),\
                                  "12th":round(grade12math.groupby(["School Name"])["math_score"].mean(),2)})

math_scores_by_grade
```

## Reading Score by Grade 

```python
#Reading scores extracted by grade
grade9reading = schooldatacomplete.loc[schooldatacomplete["grade"]=="9th","grade":"reading_score"]
grade10reading = schooldatacomplete.loc[schooldatacomplete["grade"]=="10th","grade":"reading_score"]
grade11reading = schooldatacomplete.loc[schooldatacomplete["grade"]=="11th","grade":"reading_score"]
grade12reading = schooldatacomplete.loc[schooldatacomplete["grade"]=="12th","grade":"reading_score"]

reading_scores_by_grade = DataFrame({"9th":round(grade9reading.groupby(["School Name"])["reading_score"].mean(),2),\
                                  "10th":round(grade10reading.groupby(["School Name"])["reading_score"].mean(),2),\
                                  "11th":round(grade11reading.groupby(["School Name"])["reading_score"].mean(),2),\
                                  "12th":round(grade12reading.groupby(["School Name"])["reading_score"].mean(),2)})

reading_scores_by_grade
```

## Scores by School Spending

```python
#Bins for per student spending categories
spending_bins = [0, 585, 615, 645, 675]
group_labels = ["<$585", "$585-615", "$615-645", "$645-675"]

#Bins added to above top performing schools table
top_performing_schools["School Spending"] = pd.cut(top_performing_schools["Per Student Budget"], spending_bins, labels=group_labels)

scores_by_school_spending_summary_table = DataFrame({"Average Math Score":round(top_performing_schools.groupby("School Spending")["Average Math Score"].mean(),2),\
                                                    "Average Reading Score":round(top_performing_schools.groupby("School Spending")["Average Reading Score"].mean(),2),\
                                                    "% Passing Math":round(top_performing_schools.groupby("School Spending")["% Passing Math"].mean(),2),\
                                                    "% Passing Reading":round(top_performing_schools.groupby("School Spending")["% Passing Reading"].mean(),2),\
                                                    "% Overall Passing Rate":round(top_performing_schools.groupby("School Spending")["% Overall Passing Rate"].mean(),2)})

scores_by_school_spending_summary_table
```

## Scores by School Size

```python
#Bins for school size categories
size_bins = [0, 1000, 2000, 5000]
group_names = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]

#Bins added to above top performing schools table
top_performing_schools["School Size"] = pd.cut(top_performing_schools["Total Students"], size_bins, labels=group_names)

scores_by_school_size_summary_table = DataFrame({"Average Math Score":round(top_performing_schools.groupby("School Size")["Average Math Score"].mean(),2),\
                                                    "Average Reading Score":round(top_performing_schools.groupby("School Size")["Average Reading Score"].mean(),2),\
                                                    "% Passing Math":round(top_performing_schools.groupby("School Size")["% Passing Math"].mean(),2),\
                                                    "% Passing Reading":round(top_performing_schools.groupby("School Size")["% Passing Reading"].mean(),2),\
                                                    "% Overall Passing Rate":round(top_performing_schools.groupby("School Size")["% Overall Passing Rate"].mean(),2)})
scores_by_school_size_summary_table
```
