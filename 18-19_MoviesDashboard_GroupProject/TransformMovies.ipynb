```python
#Function to correct problems with RT_ratings series
#Split as string on unwanted character, return value before character
def ratings_parse(x):
    if '%' in x:
        return x.split('%')[0]
    elif '/' in x:
        return x.split('/')[0]
    elif '–' in x:
        return x.split('–')[0]
    elif x is None:
        return None
```

```python
import pandas as pd
import numpy as np
from pandas import Series, DataFrame
import csv
import dateutil.parser
```

```python
movies = pd.read_csv('.\\output\\actor_movies_omdb.csv')
movies2 = pd.read_csv('.\\output\\actress_movies_omdb.csv')
movies.append(movies2,ignore_index = True)
```

```python
movies.head()
```

```python
#Parse Year
movies['Released'] = movies['Released'].map(lambda x: dateutil.parser.parse(x) if pd.notnull(x) else np.nan)\
                                        .map(lambda x: x.strftime('%Y-%m-%d') if pd.notnull(x) else np.nan)
movies
movies.head()
```

```python
#Normalize Rotten Tomatoes ratings
movies["RT_Ratings"] = movies["RT_Ratings"].map(lambda x: ratings_parse(x) if pd.notnull(x) else np.nan)
#Normalize imdb Ratings
movies["imdbRating"] = movies["imdbRating"].multiply(10)
```

```python
movies.head()
movies.info()
movies.columns
```

```python
movies["RT_Ratings"]=pd.to_numeric(movies["RT_Ratings"])
movies["imdbVotes"]=movies["imdbVotes"].str.replace(',', '').astype(float)
movies["BoxOffice"]=movies["BoxOffice"].str.replace('$', '')
movies["BoxOffice"]=movies["BoxOffice"].str.replace(',', '').astype(float)
#movies["BoxOffice"]=pd.to_numeric(movies["BoxOffice"])

movies = movies.rename(index=str, columns={"imdbID": "imdb_id",
                                           "imdbRating":"imdb_rating",
                                           "imdbVotes":"imdb_votes",
                                           "BoxOffice": "box_office",
                                           'Title':'title', 
                                           'Released':'released',
                                           'Runtime':'runtime', 
                                           'Genre':'genre', 
                                           'Director':'director',
                                           'Writer':'writer', 
                                           'Actors':'actores', 
                                           'Plot':'plot', 
                                           'Country':'country', 
                                           'Awards': 'awards', 
                                           'Metascore':'metascore', 
                                           'Type':'type',
                                           'Production':'production',
                                           'RT_Ratings':'rt_ratings'}) 

movies=movies.drop(columns=['Year'])

movies.info()
movies.head()
```

```python
#Combine csvs into db ready file
actors = pd.read_csv('.\\output\\actors_movies.csv')
actors2 = pd.read_csv('.\\output\\actresses_movies.csv')
#Add gender column
actors["gender"] = "male"
actors2["gender"] = "female"
all_actors = actors.append(actors2,ignore_index = True)
#Add rank
rank = pd.read_csv('.\\output\\actors.csv')
rank2 = pd.read_csv('.\\output\\actresses.csv')
rank["rank"] = rank.index +1
rank2["rank"] = rank.index +1
all_ranks = rank.append(rank2,ignore_index = True)
actordatacomplete = pd.merge(all_actors, all_ranks, how="left")
#Rename columns to SQL friendly
actordatacomplete = actordatacomplete.rename(index=str, columns={"names": "name", 
                                     "movies": "imdb_id",
                                     "credits": "credit",
                                    "ids":"actor_id"
                                    })
#Remove /name/ from actor ids
actordatacomplete['actor_id'] = actordatacomplete['actor_id'].map(lambda x: x.split('/')[1])
#Save to output
actordatacomplete.to_csv('.\\output\\actor.csv', encoding="utf-8", index=False)
```
