```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from pprint import pprint
import time
import json
import csv
from config import api_key_omdb2
```

```python
#Scrape Actors Names and IDs
html = requests.get(
    'https://www.imdb.com/list/ls022928819/').text
bs = BeautifulSoup(html,"lxml")
possible_links = bs.find_all('a')
possible_links
```

```python
names = []
ids = []
for link in possible_links:
    if link.has_attr('href'):
        if "/name/" in link.attrs['href']:
            name = link.text.replace('\n','')
            name = name.lstrip()
            id = link.attrs['href'].strip('/')
            if name not in names:
                names.append(name)
            if id not in ids:
                ids.append(id)
ids               
```

```python
actor_df = pd.read_csv('.\\output\\actors.csv')
actor_df.head()
```

```python
allmovies = []
allcredits = []
names = []
dictlist = []
ranks = []

for i,r in actor_df.iterrows():
    
        #Get Soup
        nameid = r['ids']
        name = r['names']
        rank = i + 1
        print(f"Retrieving actor number {rank} {r['names']}.")
        url = f"https://www.imdb.com/{nameid}/"
        html = requests.get(url).text
        bs = BeautifulSoup(html,"lxml")

        # Scrape data
        allidsodd = bs.find_all(class_='filmo-row odd')
        allidseven = bs.find_all(class_='filmo-row even')
        allids = allidsodd + allidseven

        #Append data      
        actordict = {}
        actordictinner = {}
        movies = []
        credits = []

        for this in allids:


            moviethisid = this.attrs['id']
            movie = moviethisid.split('-')[1]
            credit = moviethisid.split('-')[0]


            movies.append(movie)
            credits.append(credit)
            names.append(name)
            ranks.append(rank)

            actordict[name] = actordictinner
            actordictinner["rank"] = rank
            actordictinner["movies"] = movies
            actordictinner["credits"] = credits

        dictlist.append(actordict)
        allmovies.extend(movies)
        allcredits.extend(credits)      

        print(f"Retrieved actor number {rank} {name}. Total of {len(movies)} credits. Sleeping...")
        print("-----------------------------")
        time.sleep(30)
```

```python
data_tuples = list(zip(names,allmovies,allcredits))
df = pd.DataFrame(data_tuples,columns = ['names','movies','credits'])      

# df.to_csv('.\\output\\actors_movies.csv', encoding="utf-8", index=False)
df
```

```python
with open('.\\output\\actors.json', 'w') as fp:
    json.dump(dictlist, fp)
```

```python
#Load in csv with IMBD ids
omdb_link_file = pd.read_csv('.\\output\\actors_movies.csv')
omdb_link_data = pd.DataFrame(omdb_link_file)

#Add leading zeros and tt to the IMBD id column
# omdb_link_data['imdbId'] = omdb_link_data['imdbId'].apply(lambda x: 'tt{0:0>7}'.format(x))

#Convert omdb column to list
omdb_id_list = omdb_link_data['movies'].tolist()

omdb_id_list = list(dict.fromkeys(omdb_id_list))
len(omdb_id_list)
```

```python
#Set up for call
url = "http://private.omdbapi.com/"
params = {"apikey":api_key_omdb2}
not_found = []

#Set up loop
start_time = time.time()
for x in range (len(omdb_id_list)):
    print(f"Retrieving movie {x + 1} of {len(omdb_id_list)}")
    
    #movie paramater is selected from list
    movieid = omdb_id_list[x]
    params["i"] = movieid
    
    response = requests.get(url,params=params).json()
    try:
        # assemble url and make API request
        omdb_link_data.loc[x,'Title'] = response.get("Title",'')
        omdb_link_data.loc[x,'Year'] = response.get("Year",'')
        omdb_link_data.loc[x,'Released'] = response.get("Released",'')
        omdb_link_data.loc[x,'Runtime'] = response.get("Runtime",'')
        omdb_link_data.loc[x,'Genre'] = response.get("Genre",'')
        omdb_link_data.loc[x,'Director'] = response.get("Director",'')
        omdb_link_data.loc[x,'Writer'] = response.get("Writer",'')
        omdb_link_data.loc[x,'Actors'] = response.get("Actors",'')
        omdb_link_data.loc[x,'Plot'] = response.get("Plot",'')
        omdb_link_data.loc[x,'Country'] = response.get("Country",'')
        omdb_link_data.loc[x,'Awards'] = response.get("Awards",'')
        omdb_link_data.loc[x,'RT_Ratings'] = response["Ratings"][1].get("Value",'')
        omdb_link_data.loc[x,'Metascore'] = response.get("Metascore",'')
        omdb_link_data.loc[x,'imdbRating'] = response.get("imdbRating",'')
        omdb_link_data.loc[x,'imdbVotes'] = response.get("imdbVotes",'')
        omdb_link_data.loc[x,'imdbID'] = response.get("imdbID",'')
        omdb_link_data.loc[x,'Type'] = response.get("Type",'')
        omdb_link_data.loc[x,'BoxOffice'] = response.get("BoxOffice",'')
        omdb_link_data.loc[x,'Production'] = response.get("Production",'')
        
    except (KeyError, ValueError) as e:
        print(f"skipping {movieid}")
        not_found.append(movieid)
        
    #If there is no RT_rating
    except IndexError:
        omdb_link_data.loc[x,'Title'] = response.get("Title",'')
        omdb_link_data.loc[x,'Year'] = response.get("Year",'')
        omdb_link_data.loc[x,'Released'] = response.get("Released",'')
        omdb_link_data.loc[x,'Runtime'] = response.get("Runtime",'')
        omdb_link_data.loc[x,'Genre'] = response.get("Genre",'')
        omdb_link_data.loc[x,'Director'] = response.get("Director",'')
        omdb_link_data.loc[x,'Writer'] = response.get("Writer",'')
        omdb_link_data.loc[x,'Actors'] = response.get("Actors",'')
        omdb_link_data.loc[x,'Plot'] = response.get("Plot",'')
        omdb_link_data.loc[x,'Country'] = response.get("Country",'')
        omdb_link_data.loc[x,'Awards'] = response.get("Awards",'')
        omdb_link_data.loc[x,'Metascore'] = response.get("Metascore",'')
        omdb_link_data.loc[x,'imdbRating'] = response.get("imdbRating",'')
        omdb_link_data.loc[x,'imdbVotes'] = response.get("imdbVotes",'')
        omdb_link_data.loc[x,'imdbID'] = response.get("imdbID",'')
        omdb_link_data.loc[x,'Type'] = response.get("Type",'')
        omdb_link_data.loc[x,'BoxOffice'] = response.get("BoxOffice",'')
        omdb_link_data.loc[x,'Production'] = response.get("Production",'')     

end_time = time.time()
elapsed = end_time - start_time
print(f"{elapsed} seconds elapsed, {len(not_found)} record(s) not found")
```
